### Q1.] How to check wheater there is any relation between data of Song_Hits and Song_Rating?

```python
sns.set_style('whitegrid')
plt.plot(plot_Rating,plot_Hits)
plt.xlabel('Song_Rating')
plt.ylabel('Song_Hits');

plt.title("Popularity of Hits and Rating of songs")
plt.legend(['Song_Rating', 'Song_Hits']);
```

### Q3.] How to implement sns by using it plot a bar chart which shows the hits taken by artists. 

```python
sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 1, figsize=(12,8))
plt.tight_layout(pad=5);

axes[0].bar(High_Rated_Songs.Artist, High_Rated_Songs.index)
axes[0].set_title(' Hit Songs vr Artists')
axes[0].set_xlabel('Artists')
axes[0].set_ylabel('Number of Artists')


axes[1].bar(High_Rated_Songs.Song_Hits, High_Rated_Songs.index, color='orange');
axes[1].set_xlabel('Hit Songs')
axes[1].set_ylabel('Index');
```

This bar char is showing relation between hit songs and th song created by the artists.

### Q4.] By using scatterplot can we describe the page pattern followed by the Highly hited and rated Songs?

```python
plt.figure(figsize=(12, 6))
plt.title('Most Used Page_Type')

sns.scatterplot(x=High_Rated_Songs.Song_Rating, 
                y=High_Rated_Songs.Song_Hits, 
                hue=High_Rated_Songs.Page_Type,
                s=100);
```

This is one more kind of graphs called as scatterplot showing mostly used type of page by top artists. By looking into it we can able to understand Chords are mostly used in this scenario.
